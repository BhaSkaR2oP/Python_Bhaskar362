```python
#4. Dictionary Data Type
```

```python
#A dictionary stores data as key-value pairs inside {}.

#Keys are unique, values can be duplicated.
```

```python
student = {"name": "Aman", "age": 22, "grade": "A"}

print(student["name"])
```

```python
print(student.get("age"))   
```

```python
print(student.keys())       # dict_keys(['name', 'age', 'grade'])
```

```python
print(student.values()) 
```

```python
#Dictionary Manipulation
```

```python
student["age"] = 23       # update
student["city"] = "Delhi" # add new key
del student["grade"]      # delete key

for key, value in student.items():
    print(key, ":", value)
```

```python
#Example 1 – Count word frequency in a string
```

```python
text = "python is easy and python is powerful"
words = text.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```

```python
#Example 2 – Reverse words in a string
```

```python
sentence = "I love Python"
words = sentence.split()
reversed_sentence = " ".join(words[::-1])
print(reversed_sentence)
```

```python
#Example 3 – Find max number from a list
```

```python
numbers = [10, 25, 7, 33, 18]
print("Max:", max(numbers))
print("Min:", min(numbers))
print("Sum:", sum(numbers))
```
